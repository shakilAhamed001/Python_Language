1. **capitalize():**  Converts the first character of the string to uppercase.

```python
text = "hello world"
print(text.capitalize())
```

2. **casefold():** Converts the entire string to lowercase (stronger than lower()).

```python
text = "HELLO"
print(text.casefold())
```

3. **center():** Returns a centered string with a specified width.

```python
text = "Python"
print(text.center(20))
```

4. **count():** Returns the number of times a value appears in the string.

```python
text = "banana"
print(text.count("a"))
```

5. **encode():** Converts the string into encoded bytes.

```python
text = "hello"
print(text.encode())
```

6. **endswith():** Returns True if the string ends with a specified value.

```python
text = "hello.py"
print(text.endswith(".py"))
```

7. **expandtabs():** Sets the tab size of a string.

```python
text = "H\te\tl\tl\to"
print(text.expandtabs(4))
```

8. **find():** Searches for a value and returns its position.

```python
text = "hello world"
print(text.find("world"))
```

9. **format():** Formats values into a string.

```python
txt = "My name is {}"
print(txt.format("Rahim"))
```

10. **format_map():** Formats string values using a dictionary.

```python
data = {"name":"Rahim"}
print("My name is {name}".format_map(data))
```

11. **index():** Searches for a value and returns its position (error if not found).

```python
text = "hello world"
print(text.index("world"))
```

12. **isalnum():** Returns True if all characters are letters or numbers.

```python
text = "abc123"
print(text.isalnum())
```

13. **isalpha():** Returns True if all characters are alphabet letters.

```python
text = "Python"
print(text.isalpha())
```

14. **isascii():** Returns True if all characters are ASCII characters.

```python
text = "hello"
print(text.isascii())
```

15. **isdecimal():** Returns True if all characters are decimal numbers.

```python
text = "123"
print(text.isdecimal())
```

16. **isdigit():** Returns True if all characters are digits.

```python
text = "123"
print(text.isdigit())
```

17. **isidentifier():** Checks if the string is a valid Python variable name.

```python
text = "variable1"
print(text.isidentifier())
```

18. **islower():** Returns True if all characters are lowercase.

```python
text = "hello"
print(text.islower())
```

19. **isnumeric():** Returns True if all characters are numeric.

```python
text = "12345"
print(text.isnumeric())
```

20. **isspace():** Returns True if all characters are spaces.

```python
text = "   "
print(text.isspace())
```

21. **istitle():** Returns True if the string follows title case.

```python
text = "Hello World"
print(text.istitle())
```

22. **isupper():** Returns True if all characters are uppercase.

```python
text = "HELLO"
print(text.isupper())
```

23. **join()**: Joins elements of an iterable using a string separator.

```python
words = ["I","love","Python"]
print(" ".join(words))
```

24. **ljust():** Left-aligns the string in a specified width.

```python
text = "Python"
print(text.ljust(10,"-"))
```

25. **lower():** Converts all characters to lowercase.

```python
text = "HELLO"
print(text.lower())
```

26. **lstrip():** Removes whitespace from the left side.

```python
text = "   hello"
print(text.lstrip())
```

27. **maketrans():** Creates a translation table for translate().

```python
text = "hello"
table = str.maketrans("h","j")
print(text.translate(table))
```

28. **partition():** Splits the string into three parts using a separator.

```python
text = "hello-world"
print(text.partition("-"))
```

30. **replace():** Replaces one value with another.

```python
text = "I like Java"
print(text.replace("Java","Python"))
```

31. **rfind():** Returns the last occurrence position of a value.

```python
text = "banana"
print(text.rfind("a"))
```

33. **rjust():** Right-aligns the string in a specified width.

```python
text = "Python"
print(text.rjust(10,"-"))
```

34. **rpartition():** Splits string into three parts from the right side.

```python
text = "hello-world-python"
print(text.rpartition("-"))
```

35. **rsplit():** Splits the string from the right side.

```python
text = "apple,banana,orange"
print(text.rsplit(",",1))
```

36. **rstrip():** Removes whitespace from the right side.

```python
text = "hello   "
print(text.rstrip())
```

37. **split():** Splits the string into a list.

```python
text = "apple banana orange"
print(text.split())
```

38. **splitlines():** Splits the string at line breaks.

```python
text = "hello\nworld"
print(text.splitlines())
```

39. **startswith():** Returns True if the string starts with a specified value.

```python
text = "python programming"
print(text.startswith("python"))
```

40. **strip():** Removes whitespace from both sides.

```python
text = "   hello   "
print(text.strip())
```

41. **swapcase():** Converts lowercase to uppercase and vice versa.

```python
text = "Hello World"
print(text.swapcase())
```

42. **title():** Converts the first letter of each word to uppercase.

```python
text = "hello world"
print(text.title())
```

43. **translate():** Replaces characters using a translation table.

```python
text = "hello"
table = str.maketrans("e","a")
print(text.translate(table))
```

44. **upper():** Converts all characters to uppercase.

```python
text = "hello"
print(text.upper())
```

45. **zfill():** Adds zeros at the beginning of the string.

```python
text = "25"
print(text.zfill(5))
```
